# PCA

```python
#Improting the PCA module
from sklearn.decomposition import PCA
pca = PCA(svd_solver='randomized', random_state=42)
```

```python
#standardizing x
from sklearn.preprocessing import StandardScaler
df_std=StandardScaler().fit_transform(df_abs)
```

```python
#Doing the PCA on the train data
pca.fit(df_std)
```

```python
pca.components_
```

```python
colnames = list(df_abs.columns)
pcs_df = pd.DataFrame({'PC1':pca.components_[0],'PC2':pca.components_[1], 'Feature':colnames})
pcs_df.head()
```

```python
pca.explained_variance_ratio_
```

```python
#Making the screeplot - plotting the cumulative variance against the number of components
%matplotlib inline
fig = plt.figure(figsize = (12,8))
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance')
plt.show()
```

# 12 features are explaining 95% variance hence building clusters using first 12 features

# Clustering

```python
from sklearn.cluster import KMeans
from scipy.stats import zscore
```

```python
x = df_abs.drop(columns=['Education', 'Son', 'Social drinker',
       'Social smoker', 'Pet', 'Weight', 'Height', 'Body mass index',
       'Absenteeism time in hours'])
```

```python
Y = df_abs['Absenteeism time in hours']
```

```python
#scaling reduced independent variable 
df_scaled=x.apply(zscore)
```

```python
cluster_range=range(1,15)
cluster_errors=[]
for num_cluster in cluster_range:
    clusters=KMeans(num_cluster, n_init=10)
    clusters.fit(df_scaled)
    labels=clusters.labels_
    centroids=clusters.cluster_centers_
    cluster_errors.append(clusters.inertia_)
clusters_df=pd.DataFrame({"num_clusters":cluster_range, "cluster_errors":cluster_errors})
clusters_df[0:15]
```

```python
#elbow plot
plt.figure(figsize=(12,6))
plt.plot(clusters_df.num_clusters, clusters_df.cluster_errors, marker="o")
```

# 6 optimum number of clusters

```python
#model fitting  : 
kmeans=KMeans(n_clusters=6, n_init=15, random_state=2345)
kmeans.fit(df_scaled)
centroids=kmeans.cluster_centers_
centroid_df=pd.DataFrame(centroids, columns=list(df_scaled))
kmeans.labels_
pd.DataFrame(kmeans.labels_, columns=["label"]).head()
```

# Creating a new dataframe only for labels and converting it into categorical variable

```python
df_labels=pd.DataFrame(kmeans.labels_, columns=list(['labels']))
df_labels['labels']=df_labels["labels"].astype("category")
```

# #joining the label dataframe with the original dataframe

```python

abs_df_labeled=df_abs.join(df_labels)
df_analysis=(abs_df_labeled.groupby(['labels'],axis=0)).head()
df_analysis
```
