```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
#to show matplot every time
%matplotlib inline
```

```python
n=7
```

```python
p=.6
```

```python
k=np.arange(0,8)
```

```python
k
```

```python
binomial = stats.binom.pmf(k,n,p)
```

```python
binomial
```

plt.plot(k,binomial,'o-')

```python
plt.plot(k,binomial,'o-')
plt.title('binomial: n=%i , p = %.2f'% (n,p) , fontsize = 15)
plt.xlabel('Number of Successes')
plt.ylabel('Probability of successes')
plt.show()
```

```python
lam = 3
```

```python
n=np.arange(0,15)
```

```python
poison=stats.poisson.pmf(n,lam)
```

```python
poison
```

```python
plt.plot(n,poison,'o-')
plt.title('poison: $/lamda$ = %i' %lam)
plt.xlabel('Number of customers arriving in a minute ')
plt.ylabel('probability of customers coming in one minute')
```
